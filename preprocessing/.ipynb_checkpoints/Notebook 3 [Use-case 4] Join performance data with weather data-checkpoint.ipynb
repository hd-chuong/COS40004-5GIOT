```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import geopy
import time
from tqdm.notebook import tqdm_notebook
import datetime
from datetime import datetime
import itertools
%matplotlib inline
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})
```

```python
minute_data = pd.read_csv("E:/Y5/Topics/data/datasets/app-dependent/minute.csv").copy()
tqdm_notebook.pandas()
minute_data["date_format"] = minute_data.progress_apply(lambda r: str(datetime(int(r["Year"]), int(r["Month"]), int(r["Date"]))), axis=1)
minute_data
```

```python
weather_data = pd.read_csv("E:/Y5/Topics/data/datasets/app-dependent/weather-data.csv", parse_dates=False, dtype=str).copy()
weather_data['date_format'] = weather_data.progress_apply(lambda r: str(datetime.strptime(r.Date, '%d/%m/%Y')), axis=1)
weather_data = weather_data.drop(columns=["Minimum temperature (°C)", "Maximum temperature (°C)", "Month", "Date"])
weather_data = weather_data.rename(columns={
    'Numerical data - Rain': 'Rain Rate', 
    'Numerical data - Cloud': 'Cloud Rate',
    'Numerical data - Wind' : 'Wind Rate'})
weather_data
```

```python
final_data = minute_data.merge(weather_data, on="date_format", how="left")
final_data.columns
```

```python
final_data.to_csv("E:/Y5/Topics/data/datasets/app-dependent/date_and_weather.csv", index=False)
```
