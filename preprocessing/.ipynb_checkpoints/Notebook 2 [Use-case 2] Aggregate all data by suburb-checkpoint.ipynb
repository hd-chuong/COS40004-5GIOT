# 5G-IOT visual interface: Translate GPS location into suburbs and postcodes

The aim of this notebook is to generate the GEOJSON file that contains the geometrical boundary and the average 5G performance attributes of each suburb in Brimbank.

We use the minute data which has enough resolution for suburb-wise aggregation. The geoboundary dataset is needed for suburb translation pipeline. The geoboundary dataset can be download from [here](https://github.com/tonywr71/GeoJson-Data/blob/master/suburb-2-vic.geojson)

The output GEOJSON file is ready for **Use case 2. Compare 5G connectivity amongst suburbs**



Note that this notebook uses GeoPandas. There might be some overheads when installing Geopandas module. It is recommended to install a virtual environment that isolates the GeoPandas package from other Python packages that are local to your computer.

```python
import geopandas
from shapely.geometry import Point
import pandas as pd
import geopy
from tqdm.notebook import tqdm_notebook
from tqdm import tqdm_notebook as tqdm
```

### Define constants

```python
base = "../data/sources/"
output_folder = "../data/outputs/"

BRIMBANK_SUBURBS = [
 'ALBANVALE',
 'ALBION',
 'ARDEER',
 'BROOKLYN',
 'CAIRNLEA',
 'CALDER PARK',
 'DEER PARK',
 'DELAHEY',
 'DERRIMUT',
 'KEALBA',
 'KEILOR',
 'KEILOR DOWNS',
 'KEILOR EAST',
 'KEILOR LODGE',
 'KEILOR NORTH',
 'KEILOR PARK',
 'KINGS PARK',
 'ST ALBANS',
 'SUNSHINE',
 'SUNSHINE NORTH',
 'SUNSHINE WEST',
 'SYDENHAM',
 'TAYLORS LAKES',
 'TULLAMARINE'
]
```

### View the minutely dataset file

```python
minute_data = pd.read_csv(f"{output_folder}/minute.csv")
minute_data
```

### View the geoboundary data

```python
suburb_geodata = geopandas.read_file(f'{base}/suburb-2-vic.geojson.txt')
suburb_geodata.plot()
```

### Translate GPS location into suburbs

```python
def translate_to_suburb(lookup_table, row):
    point = Point(row['longitude'], row['latitude'])
    matched = lookup_table.geometry.contains(point)
    matched = matched[matched == True]
    if len(matched) == 0:
        return -1
    index = matched.index[0]
    return lookup_table.loc[index].vic_loca_2
```

```python
minute_data_with_suburb = minute_data.copy()
tqdm_notebook.pandas()
minute_data_with_suburb["suburb"] = minute_data_with_suburb.progress_apply(lambda r: translate_to_suburb(suburb_geodata, r), axis=1)
```

At this step, we already identified the suburb name for each data point. It's time to perform calculate the average 5G connectivity for each suburb.

## Performance the aggregation per suburb

First we need to define the aggregation method, which is mean for every performance attributes


```python
performance_columns = {
    'Upload speed (Mbits/sec)': ['mean'],
    'Download speed (MBits/sec)': ['mean'],
    'Download size (MBytes)': ['mean'],
    'Upload size (MBytes)': ['mean'],
    'Upload throughput (MBits)': ['mean'],
    'Response time from CloudflareDNS (ms)': ['mean'],
    'Response time from GoogleDNS (ms)': ['mean'],
    'Response time from OptusDNS (ms)': ['mean'],
    'Response time from EC2DNS (ms)': ['mean']
}
```

```python
def aggregate_by_suburb(data):
    groupby_column = "suburb"
    
    aggregated = data.copy()
    aggregated = aggregated.groupby([groupby_column], as_index=False).agg(performance_columns)
    aggregated.columns = aggregated.columns.get_level_values(0)
    
    aus_poas = geopandas.read_file(f'{base}/suburb-2-vic.geojson.txt')
    aus_poas = aus_poas[aus_poas["vic_loca_2"].isin(BRIMBANK_SUBURBS)]
    aus_poas = aus_poas[["vic_loca_2", "geometry"]]
    aus_poas[groupby_column] = aus_poas.apply(lambda r: r["vic_loca_2"], axis=1)
    merged = aus_poas.merge(aggregated, on=groupby_column, how="left")
    return merged
    

grouped_by_suburb = aggregate_by_suburb(minute_data_with_suburb)
grouped_by_suburb.to_file(output_folder + "/suburb-performance.json", driver="GeoJSON") 
```
